```python
def create_spiral_surface(A, B, n=1000, size=(500, 500)):
    t = np.linspace(0, 4 * np.pi, n)
    x = A * t * np.cos(B * t)
    y = A * t * np.sin(B * t)

    fig1 = plt.figure()
    ax = fig1.add_subplot()
    ax.plot(x, y, linewidth=5)
    
    x -= np.min(x)
    y -= np.min(y)
    
    fig2 = plt.figure()
    ax = fig2.add_subplot()
    ax.plot(x, y, linewidth=5)

    x_pixels = np.expand_dims(np.asarray((x / np.max(x)) * size[0] - 1, dtype=np.int16), axis=1)
    y_pixels = np.expand_dims(np.asarray((y / np.max(y)) * size[1] - 1, dtype=np.int16), axis=1)
    ax.scatter(x_pixels,y_pixels, color='red', marker='o', s=2)
    print(x_pixels)
    
    coords = np.concatenate([x_pixels, y_pixels], axis=1)
    midpoint = np.array(np.floor((size[0] - 1) / 2), dtype=np.int16)
    delta = (midpoint - coords[0, :]) * 2
    coords += delta
    print(size, pg.Vector2(delta[0], delta[1]))
    
    # spiral_surf = Screen(size + pg.Vector2(*delta), font=None)
    # for pos in coords:
    #     spiral_surf.update_pixels(pos, base=True)
    #     
    # spiral_surf.refresh()
    # 
    # spiral_array = pg.surfarray.pixels3d(spiral_surf.surface)
    # alpha_channel = np.expand_dims(pg.surfarray.array_alpha(spiral_surf.surface), axis=2)
    # spiral_array = np.concatenate([spiral_array, alpha_channel], axis=2)
    # return spiral_surf
```

```python
def varying_intervals(x):
    return np.exp(0.7 * x)  # Example varying interval function

# Define the start and stop points
start_point = 0
stop_point = 100

# Define the number of intervals
num_intervals = 100

# Generate points with varying intervals
points = np.cumsum(varying_intervals(np.linspace(0, 1, num_intervals)))
print(points)
# Plot the points
plt.plot(points, np.zeros_like(points), marker='o')
plt.title('Array with Increasing Intervals')
plt.xlabel('Points')
plt.ylabel('Zero Line')
plt.grid(True)
plt.show()
```

```python


# Replace 'your_file.csv' with the path to your CSV file
os.chdir('/Users/Thinkpad/Desktop/Warwick/hero-monitor')

# Read the CSV file into a Pandas DataFrame with additional parameters
spiraldata = pd.read_csv('spiraldata.csv')


def radial_speed(df):
    plt.plot(df['time'], abs(df['theta']))
    plt.title('Radial Speed Over Time')
    plt.xlabel('Time')
    plt.ylabel('Radial Speed (|Theta|)')
    plt.grid(True)
    plt.show()

def cum_dist(df):
    df['dist'] = np.linalg.norm(df[['rel_pos_x', 'rel_pos_y']].diff(axis=0), axis=1)
    df['cum_dist'] = df['dist'].cumsum()
    plt.plot(df['time'], df['cum_dist'])
    plt.title('Cumulative Distance Over Time')
    plt.xlabel('Time')
    plt.ylabel('Cumulative Distance')
    plt.grid(True)
    plt.show()
    
def error(df):
    plt.plot(df['time'], df['error'])
    plt.title('Error Over Time')
    plt.xlabel('Time')
    plt.ylabel('Error')
    plt.grid(True)
    plt.show()
    
    

```

```python
# Initial interval size
initial_interval = 2

# Final interval size
final_interval = 0.1

# Number of elements for each interval
elements_per_interval = 10

# Calculate the number of intervals
num_intervals = int((initial_interval - final_interval) / 0.1) + 1

# Initialize an empty list to store subarrays
arrays = []

# Generate arrays with variable intervals
for i in range(num_intervals):
    interval_size = initial_interval - i * 0.1
    subarray = np.linspace(i * elements_per_interval, (i + 1) * elements_per_interval - 1, elements_per_interval)
    subarray *= interval_size
    arrays.append(subarray)

# Concatenate the subarrays into a single array
result_array = np.concatenate(arrays)

# Print the result array
print(result_array)
```
